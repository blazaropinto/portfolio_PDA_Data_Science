```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data from https://usmart.io/org/cyclingscotland/discovery/discovery-view-detail/f1e8a739-ee14-4742-a7de-45d1ab6cb8aa

df = pd.read_csv('National Monitoring Framework - all-mode traffic survey results Sept 2020 - Cycling Scotland.csv',
                index_col='ID')
df.columns
```

```python
df.head()
```

```python
transport =['Car', 'Taxi', 'LGV', 'HGV', 'Service Bus', 'Coach',
       'MCycle', 'Cyclist', 'Pedestrian']

for m in transport:
    print(m, '\n', df[m].value_counts(),'\n')
```

```python
df.replace(['No data', 'No Data', '.'], np.nan, inplace=True)

for m in transport:
    print(m, '\n', df[m].value_counts(),'\n')
```

```python
for t in transport:
    df[t] = df[t].astype('float64')
```

```python
by_site = df.groupby(['Local Authority', 'Latitude', 'Longitude'])['Car', 'Taxi', 'LGV', 'HGV', 'Service Bus', 'Coach',
       'MCycle', 'Cyclist', 'Pedestrian'].sum()
by_site
```

```python
by_site['Vehicle with emissions'] = by_site.Car + by_site.Taxi + by_site.LGV + \
                                    by_site.HGV + by_site.Coach + by_site['Service Bus']
by_site['Active Transport'] = by_site.MCycle + by_site.Cyclist + by_site.Pedestrian
by_site = by_site[['Vehicle with emissions', 'Active Transport']]
by_site.head()
```

```python
by_site = by_site.reset_index()
by_site.head()
```

```python
at_med = by_site['Active Transport'].median()
```

```python
import folium
```

```python
scotmap = folium.Map(location=[55.934206343443066, -3.246200181790759], zoom_start=7, tiles='OpenStreetMap')
```

```python
def scoticon(at):
    if at > at_med:
        color='green'
    else:
        color='lightgray'
    return color
```

```python
html = """
<h4> {}
</h4> {}
"""

layer = folium.FeatureGroup(name='Active Transport')
```

```python
for i, row in by_site.iterrows():
    iframe = folium.IFrame(html=html.format('Active Transport: ' + str(row[4]), 
                                            'Vehicle with emissions: ' + str(row[3])), width=250, height=80)
    layer.add_child(folium.Marker(location=[row[1],row[2]], 
                                    popup=folium.Popup(iframe), icon=folium.Icon(color=scoticon(row[4]))))
```

```python
scotmap.add_child(layer)
```

```python
scotmap.save('example_scotmap.html')
```
